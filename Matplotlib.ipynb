```python
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
#w = np.tan(x)
plt.plot(x,y,color='green')
plt.plot(x,z,color='red')
#plt.plot(x,w)
```

```python
fig,ax = plt.subplots()
ax.plot(x,y,color='red')
ax.fill_between(x,y,color='green',alpha=0.3)
```

```python
x = np.linspace(-1,1,100)
y = np.exp(x) # .abs() , .log()
plt.scatter(x,y,color='red')
```

```python
plt.scatter(x,y,c=x)
```

```python
# line graph

a = [1,2,3,4,5,6,7]
b = [2 ,4 ,6 ,8 , 10,12,14]
c = (1,2,4,8,16,32,64)
ax = plt.axes()
ax.set_facecolor("pink")
plt.title("Line Graph")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(a,b,color='red' , label='line 1',linewidth = 2)
plt.plot(a,c,color='green' , label='line 2', linestyle = 'dashed' , marker = '*', markersize = 10, markerfacecolor='blue',markeredgecolor='red')
# solid dashed dotted dashdot - linestyles
# * . o ^ > < v and so on...
plt.legend()
```

```python
# Bar Graph
x = ['A','B','C','D','E']
y = [30,35,20,15,50]
plt.title("Simple Bar Graph",fontsize=12)
colr = ['red','green','blue','yellow']
plt.xticks(rotation = 45)
plt.grid()
plt.bar(x,y,color = colr)
plt.savefig('bar_graph.png')
```

```python
x = ['A','B','C','D','E']
y = [30,35,20,15,50]
plt.title("Simple Bar Graph",fontsize=12)
colr = ['red','green','blue','yellow']
plt.xticks(rotation = 45)
plt.grid()
plt.barh(x,y,color = colr)
plt.savefig('bar_graph.png')
```

```python
# Pie Chart
# air compostion
x = ['Nitrogen','Oxygen','CO2','Hydrogen','Others']
values = [57,21,10,13,4]
# plt.figure(figsize=(15,15))
colrs = ['red','yellow','green','cyan','purple']
#colrs = ['r','g','b','k','c']
#colrs = ['#FF5733','#FF5733','#FF5733','#FF5733','#FF5733']
exp1 = [0,0,0.2,0,0]
plt.pie(values,labels=x,explode=exp1,autopct='%1.1f%%',shadow = True,colors = colrs, counterclock=False, startangle = 90)
plt.legend(loc='lower right')
plt.title("Air Composition")
plt.show()
```

```python
# Histogam
x = np.random.normal(10,15,250)
plt.hist(x,histtype='stepfilled')
plt.show()
```
