```python
tips = sns.load_dataset('tips')
# data - ordered collection of data which can be used for analytics and visualizataion purpose
tips
```

```python
tips.shape
```

```python
# distplot
# distributional plot
sns.distplot(tips['total_bill'],hist=False)
```

```python
sns.distplot(tips['total_bill'],kde=False) # kde = kernal density estimate
```

```python
import matplotlib.pyplot as plt
plt.scatter(tips['total_bill'],tips['tip'])
# sns.scatterplot(x='total_bill',y='tip',hue='day',data=tips,alpha=0.8,s=70)
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips,alpha=0.8,s=70)
```

```python
# relational plot
#sns.relplot(x='total_bill',y='day',data=tips,hue='day',style='smoker')
sns.relplot(x='total_bill',y='day',data=tips,hue='size',style='smoker')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='day',col='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='day',col='time',row='sex')
```

```python
# join plot 
# universal distrirbution 
# Bivariate distribution 
sns.jointplot(x='total_bill',y='tip',data = tips,color ='red')
```

```python
sns.jointplot(x='total_bill',y='tip',data = tips,color ='red',kind = 'scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data = tips,color ='green',kind = 'reg')
# kind = scatter , kde  , resid  , reg, hex , hist
```

```python
# pairplot
sns.pairplot(tips,hue='day')
```

```python
# catagorical plotting 
# sns.stripplot
# sns.swarmplot
# sns.boxplot
# sns.boxenplot
# sns.violinplot
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(tips['total_bill'])
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxenplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# HeatMaps
import numpy as np
plt.figure(figsize = (15,10))
x = np.random.rand(10,15)
x_sort = np.sort(x)
sns.heatmap(x_sort,cmap='YlGnBu',annot=True) #cbar = False to remove color bar
```
