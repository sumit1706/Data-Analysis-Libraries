```python
def os(ls):
  if ls=='Windows 10' or ls=='Windows 10 s' or ls=='Windows 7' :
    return 'Winodws'
  elif ls == 'macOS' or ls == 'Mac OS X':
    return 'Mac'
  else:
    return 'others'
df['OS'] = df.OpSys.apply(os)
df
```
